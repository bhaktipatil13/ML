```python
df = dg.iloc[:, 0].str.split(',', expand=True)
```

```python
df.head()
```

```python
df.replace('None', np.nan, inplace=True)

df.dropna(how='all', inplace=True)

df.drop_duplicates(inplace=True)

df.fillna('NoPurchase', inplace=True)

df_onehot = pd.get_dummies(df.stack(), prefix='', prefix_sep='').groupby(level=0).sum()

df_onehot.head()

```

```python

df_onehot = df_onehot.astype(bool).astype(int)

from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df_onehot, min_support=0.05, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# Display results
print("Frequent Itemsets:\n", frequent_itemsets)
print("\nAssociation Rules:\n", rules)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df_onehot, min_support=0.03, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

rules = rules[rules["lift"] > 1]

rules_sorted = rules.sort_values(by="lift", ascending=False)

print("Frequent Itemsets:\n", frequent_itemsets)

print("\nAssociation Rules:\n", rules_sorted)
```

```python
""" Key Metrics
Each rule consists of:

Antecedent (LHS) → the item(s) already purchased.
Consequent (RHS) → the item(s) likely to be purchased as a result.
Support → how frequently the itemset appears in transactions.
Confidence → how often the consequent is purchased when the antecedent is purchased.
Lift → how much more likely the consequent is bought compared to chance (Lift > 1 means strong association)."""

```
